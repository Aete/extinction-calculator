```python
import os

import numpy as np
import pandas as pd
```

```python
raw_data_dir = './raw_data/future'
list_csv = os.listdir(raw_data_dir)
list_working_age = ['15-19세','20-24세',
       '25-29세', '30-34세', '35-39세', '40-44세', '45-49세', '50-54세',
       '55-59세', '60-64세'  ]

dict_sido = {
    'seoul':'서울',
    'busan':'부산',
    'daegu':'대구',
    'incheon':'인천',
    'gwangju':'광주',
    'daejeon':'대전',
    'ulsan':'울산',
    'sejong':'세종',
    'gyeonggi':'경기',
    'gangwon':'강원',
    'chungbuk':'충북',
    'chungnam':'충남',
    'jeonbuk':'전북',
    'jeonnam':'전남',
    'gyeongbuk':'경북',
    'gyeongnam':'경남',
    'jeju':'제주'
}
```

```python
def process_data(csv):

    df_sample = pd.read_csv(f'{raw_data_dir}/{csv}')       
    df_sample['sido'] = dict_sido[csv.replace('_future_pop.csv','')]
    df_sample.drop('성별', axis=1, inplace=True)
    df_sample.rename(columns={'시군구별':'sigungu', '연령별':'age'}, inplace=True)

    df_total = df_sample.loc[df_sample['age']=='계'].drop('age', axis=1)

    df_working_age = df_sample.loc[df_sample['age'].str.replace(' ','').isin(list_working_age)]
    df_working_age = df_working_age.groupby(['sido','sigungu'], as_index=False).sum(numeric_only=True)

    return df_total, df_working_age

list_df_total = []
list_df_working_age = []

for csv in list_csv:
    print(csv)
    df_total, df_working_age = process_data(csv)
    list_df_total.append(df_total)
    list_df_working_age.append(df_working_age)

df_total = pd.concat(list_df_total, ignore_index=True)
df_working_age = pd.concat(list_df_working_age, ignore_index=True)
```
